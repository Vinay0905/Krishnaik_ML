```python
import sqlite3
```

```python
connection=sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
cursor.execute('''
    Create table If Not Exists employees(
               if Integer Primary Key,
               name Text Not NULL,
               age Integer,
               department text

               )
''')
connection.commit()
```

```python
cursor.execute(
    '''
    Select * from employees
'''
)
```

```python
cursor.execute(

    '''
    Insert Into employees(name,age,department)
    values('Naga',21,'Ai Enginer')
'''
)
cursor.execute(

    '''
    Insert Into employees(name,age,department)
    values('Vinay',21,'Ai Enginer')
'''
)
cursor.execute(

    '''
    Insert Into employees(name,age,department)
    values('Avvaru',21,'Ai Enginer')
'''
)
connection.commit()
```
