```python
import logging
```

```python
logging.basicConfig(level=logging.DEBUG)
```

```python
logging.debug(" this is a debug message ")
logging.info(" This is a logging info message ")
logging.warning(" THis is the warnign message")
```

```python
logging.basicConfig(
    filename='App.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d  %H:%M:%S'
    )
logging.debug(" this is a debug message ")
logging.info(" This is a logging info message ")
logging.warning(" THis is the warnign message")
```

```python
logger1=logging.getLogger("Module1")
logger1.setLevel(logging.DEBUG)

logger2=logging.getLogger("Module2")
logger2.setLevel(logging.WARNING)

logging.basicConfig(
    filename='App.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d  %H:%M:%S'
    )
```

```python
logger1.debug(
    "This is debug message for module 1"
)
logger2.warning(
    "This is debug message for module 2"
    
)
```
