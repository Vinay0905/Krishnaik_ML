```python
import gc
class Myobject:
    def __init__(self,name):
        self.name=name
        print(f"Object {self.name} created")
        
    def __del__(self):
        print(f"Object {self.name} deleted")




obj1=Myobject("obj1")
obj2=Myobject("obj2")


obj1.ref=obj2
obj2.ref=obj1

del obj1
del obj2


gc.collect()
```

```python
def generate_numebrs(n):
    for i in range(n):
        yield i

for num in generate_numebrs(10000):
    print(num)
    if num>10 :
        break


```

```python
import tracemalloc

def create_list():
    return [i for i in range(1000)]

def main():
    tracemalloc.start()
    create_list()
    snapshots=tracemalloc.take_snapshot()
    top_stats=snapshots.statistics('lineno')

    print("[ Top 10]")
    for stat in top_stats[:10]:
        print(stat)
```

```python
main()
```
